```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa_data = pd.read_csv(r"E:\Kaggle Data Sets\fifa.csv", index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
plt.figure(figsize=(16,6))
plt.title("Fifa_Data(Line Plot)")
sns.lineplot(data=fifa_data)
```

```python
fifa_data.tail()
```

```python
list(fifa_data.columns)
```

```python
plt.figure(figsize=(14,6))
plt.title("Fifa_Data")
sns.lineplot(data=fifa_data['ARG'], label="ARG")
sns.lineplot(data=fifa_data['GER'], label="GER")
plt.xlabel("Date")
```

```python
plot_data=fifa_data.head(10)
plt.figure(figsize=(14,6))
plt.title("Fifa_data(Bar Plot)")
sns.barplot(x=plot_data.index, y=plot_data['ARG'])
plt.xlabel("Date")
```

```python
plot_data=fifa_data.head()
plt.figure(figsize=(14,6))
plt.title("Fifa_data(Heat Map)")
sns.heatmap(data=plot_data, annot=True)
plt.xlabel("Date")
```

```python
candy_data = pd.read_csv(r"E:\Kaggle Data Sets\candy.csv")
candy_data.head()
```

```python
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
sns.set_style("dark")
```

```python
sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
sns.set_style("dark")
```

```python
sns.scatterplot(x=candy_data['pricepercent'], y=candy_data['winpercent'], hue=candy_data['chocolate'])
sns.set_style("dark")
```

```python
sns.lmplot(x="pricepercent", y="winpercent", hue="chocolate", data=candy_data)
sns.set_style("dark")
```

```python
cancer_b_data = pd.read_csv(r"E:\Kaggle Data Sets\cancer_b.csv", index_col="Id")
cancer_m_data = pd.read_csv(r"E:\Kaggle Data Sets\cancer_m.csv", index_col="Id")
cancer_b_data.head()
cancer_m_data.head()
```

```python
plt.figure(figsize=(14,6))
sns.set_style("dark")
sns.distplot(a=cancer_b_data['Area (mean)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'], label="Malignant", kde=False)
plt.legend()
```

```python
plt.figure(figsize=(14,6))
sns.set_style("dark")
sns.kdeplot(data=cancer_b_data['Radius (worst)'], shade=True, label="Benign")
sns.kdeplot(data=cancer_m_data['Radius (worst)'], shade=True, label="Malignant")
```

```python
sns.jointplot(x=cancer_b_data['Radius (mean)'], y=cancer_b_data['Texture (mean)'], kind="kde")
sns.set_style("dark")
```

```python
sns.jointplot(x=cancer_m_data['Radius (mean)'], y=cancer_m_data['Texture (mean)'], kind="kde")
sns.set_style("dark")
```
